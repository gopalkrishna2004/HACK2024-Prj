```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/91897/Desktop/PRJ 1/Institute of Chemical Technology.csv")
df
```

```python
FIN=df.drop(['Authors','Author(s) ID','Volume','Issue','Art. No.','Page start','Page end','Page count','DOI','Link','Document Type','Publication Stage','Open Access','Source','EID'],axis=1)
```

```python
FIN.rename(columns = {'Cited by':'C'}, inplace = True)
FIN['C'].replace(np.nan,0,inplace=True)
FIN.drop(FIN[FIN['Year'] == 2024].index, inplace=True)
print(FIN)
```

```python
def impact(C, h, Year):
    k = 0.001  # Initial value of k
    while C >= (h + 1) ** k:
        k += 0.00001  # Increment 'k' by 0.01 (you can adjust this increment)
    return k
```

```python
#Apply the impact function to the DataFrame
FIN['k'] = FIN.apply(lambda row: impact(row['C'], row['h'], row['Year']), axis=1)
```

```python
# Calculate 'date' and 'kprime'
FIN['date'] = 2023 - FIN['Year']
FIN['kprime'] = FIN['k'] / FIN['date']
# Print the resulting DataFrame
print(FIN)
```
