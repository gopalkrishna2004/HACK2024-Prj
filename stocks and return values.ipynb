```python
import math
#we have converted correlation into distance and then using MDS, converted it to two dimensional coordinates
dist =pow(2*(1-c1),1/2)
from sklearn.manifold import MDS

mds = MDS(random_state=0)
scaled_df = mds.fit_transform(dist) #for two dimension
plt.scatter(scaled_df[:,0], scaled_df[:,1])

plt.show()
```

```python
from sklearn.cluster import KMeans

data = list(zip(scaled_df[:,0], scaled_df[:,1]))
inertias = []

for i in range(1,9):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,9), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

plt.scatter(scaled_df[:,0], scaled_df[:,1], c=kmeans.labels_)
for i, txt in enumerate(dist.index):
    plt.annotate(txt, (scaled_df[:,0][i], scaled_df[:,1][i]))
plt.show()
```
